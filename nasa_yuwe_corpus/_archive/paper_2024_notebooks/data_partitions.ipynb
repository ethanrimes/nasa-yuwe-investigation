```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
LANGUAGE_FILE="data/nasa_sin_cartas.csv"
LANGUAGE_FILE_TRAIN="data/nasa_sin_cartas_train.csv"
LANGUAGE_FILE_DEV="data/nasa_sin_cartas_dev.csv"
LANGUAGE_FILE_TEST="data/nasa_sin_cartas_test.csv"
```

```python
trans_df = pd.read_csv(LANGUAGE_FILE)
print(trans_df.shape)
```

```python
df_train, df_devtest = train_test_split(trans_df, test_size=0.2, random_state=42)
df_dev, df_test = train_test_split(df_devtest, test_size=0.5, random_state=42)
```

```python
print(df_train.shape)
print(df_dev.shape)
print(df_test.shape)
```

```python
df_train.to_csv(LANGUAGE_FILE_TRAIN, index=False)
df_dev.to_csv(LANGUAGE_FILE_DEV, index=False)
df_test.to_csv(LANGUAGE_FILE_TEST, index=False)
```
